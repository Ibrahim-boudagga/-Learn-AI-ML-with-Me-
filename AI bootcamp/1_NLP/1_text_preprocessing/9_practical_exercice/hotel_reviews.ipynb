```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.util import ngrams
from collections import Counter
import pandas as pd
```

```python
nltk.download("punkt_tab")
nltk.download("stopwords")
```

```python
data = pd.read_csv("tripadvisor_hotel_reviews.csv")
data.head()
```

```python
data.info()
```

```python
data["Review"][0]
```

# Lowercasing

```python
data["Review_lower"] = data["Review"].str.lower()
data.head()
```

# remove punctuation

```python
data["Review_lower_no_punct"] = data["Review_lower"].str.replace(r"[^\w\s]", "")
data.head()
```

```python
data["Review_lower_no_punct"][0]
```

# Remove StopWords

```python
en_stopwords = stopwords.words("english")
en_stopwords.remove("not")
data["Review_lower_no_punc_no_stopwords"] = data["Review_lower_no_punct"].apply(
    lambda x: " ".join([word for word in x.split() if word not in en_stopwords])
)
data.head()
```

```python
data["Review_lower_no_punc_no_stopwords"][0]
```

# Tokenization

```python
data["tokenized"] = data[
    "Review_lower_no_punc_no_stopwords"
].apply(lambda x: nltk.word_tokenize(x))

data.head()
```

```python
data["tokenized"][0]
```

# Stemming

```python
ps=PorterStemmer()
data["stemmed"]=data["tokenized"].apply(
  lambda x: " ".join([ps.stem(word) for word in x])
)
data.head()
```

```python
data["stemmed"][0]
```

# Lemmitization

```python
lm=WordNetLemmatizer()
data["lemmatized"]=data["stemmed"].apply(
  lambda x: " ".join([lm.lemmatize(word) for word in x.split()])
)
data.head()
```

```python
data["lemmatized"][0]
```

# N-grams

```python
tokens=[]
for text in data["lemmatized"]:
  tokens.extend(text.split())
tokens
```

# Unigrams

```python
unigram_tokens = Counter(ngrams(tokens, 1)).most_common(10)
unigram_tokens
```

```python
bigram_tokens = Counter(ngrams(tokens, 2)).most_common(10)
bigram_tokens
```

```python
trigram_tokens = Counter(ngrams(tokens, 3)).most_common(10)
trigram_tokens
```
