```python
import pandas as pd

bbc_data=pd.read_csv("bbc_news.csv")
bbc_data.head()
```

```python
titles=pd.DataFrame(bbc_data["title"])
titles.head()
```

# Clean Data

```python
titles["lowercase"]=titles["title"].str.lower()
titles.head()
```

```python
#remove stopwords
import nltk
en_stopwords=nltk.corpus.stopwords.words("english")
titles["stopwords"]=titles["lowercase"].apply(lambda x:' '.join(word for word in x.split() if word not in en_stopwords))
titles.head()
```

```python
#remove punctuation
import re
titles["no_punctuation"]=titles.apply(lambda x: re.sub(r'[^\w\s]',' ',x["stopwords"]),axis=1)
titles.head()
```

```python
#tokenize
from nltk.tokenize import word_tokenize
titles["tokenized_original"]=titles.apply(lambda x: word_tokenize(x["title"]),axis=1)
titles["tokenized_cleaned"]=titles.apply(lambda x: word_tokenize(x["no_punctuation"]),axis=1)
titles.head()
```

```python
# lemmatize
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
titles["tokens_clean_lemmatized"] = titles["tokenized_cleaned"].apply(
    lambda tokens: [lemmatizer.lemmatize(token) for token in tokens]
)
titles.head()
```

```python
tokensOriginalList=sum(titles["tokenized_original"],[])#flatten the list
tokensCleanedList=sum(titles["tokenized_cleaned"],[])#flatten the list
```

# POS

```python
import spacy
nlp=spacy.load("en_core_web_sm")

doc=nlp(" ".join(tokensOriginalList))
```

```python
pos_rows = [{
  "token": token.text, 
  "pos": token.pos_, 
  "dep": token.dep_,
} for token in doc
]
pos_df = pd.DataFrame(pos_rows)
```

```python
pos_df_counts = (
    pos_df.groupby(["token", "pos"])
    .size()
    .reset_index(name="counts") #type: ignore
    .sort_values(by="counts", ascending=False)
)
pos_df_counts.head()
```

```python
# most common nouns
nouns = pos_df_counts[pos_df_counts.pos == "NOUN"][:10]
nouns
```

```python
#most common verbs
verbs = pos_df_counts[pos_df_counts.pos == "VERB"][0:10]
verbs
```
