# TF-IDF

TF-IDF is a method of vectorizing text data.

It is a good method for text data that is not too large.

TF is the term frequency of a word in a document.

IDF is the inverse document frequency of a word in the corpus.

TF-IDF is better than Bag of Words because it is a more accurate method of vectorizing text data.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
data=[
    " Most shark attacks occur about 10 feet from the beach since that is where the people are",
    "the efficiency with which he paired the socks in the drawer was quite admirable",
    "carol drank the blood as if she were a vampire",
    "giving directions that the mountains are to the west only works when you can see them",
    "the sign said there was road work ahead so he decided to speed up",
    "the gruff old man sat in the back of the bait shop grumbling to himself as he scooped out a handful of worms",
]
```

```python
vectorizer=TfidfVectorizer()
```

```python
data_tfidf=vectorizer.fit_transform(data)
```

```python
pd_data=pd.DataFrame(
  data_tfidf.toarray(), # type: ignore
  columns=vectorizer.get_feature_names_out(),
)
pd_data
```

every word in the document is a column in the dataframe.

the value of the word in the document is the tf-idf value of the word in the document,

which means that the word that has the highest tf-idf value in the document is the most important word in the document.
